```python
import pandas as pd

df_employees = pd.read_csv ('../Project-Anaconda/employees.csv')
df_system_access = pd.read_csv ('../Project-Anaconda/system_access.csv')

```

```python
df_system_access

df_system_access['id'] = 'EMP' + df_system_access['id'].astype(str).str.zfill(6)
df_system_access
```

```python
df_system_access.rename(columns={'id':'employee_number'}, inplace=True)
df_system_access.set_index('employee_number', inplace=True)
df_employees = df_employees.set_index('employee_number')
df = df_employees.join(df_system_access, how='left')
df.head(15)
```

```python
df.loc[df['access_level'].isnull()]
```

```python
subset_df_system_access = df.loc[df['access_level'].isnull()]
```

```python
subset_df_system_access.loc[subset_df_system_access['title'] == 'Account Manager', 'access_level'] = 'Approval - Manager'
subset_df_system_access.tail()
```

```python
df_missingAL = df.loc[df['access_level'].isnull()]
df_missingAL
```

```python
subset_df_system_access.loc[subset_df_system_access['title'] == 'Sales Assistant', 'access_level'] = 'Contract Entry'
subset_df_system_access.tail()
```

```python
subset_df_system_access.loc[subset_df_system_access['title'] == 'Director of Sales', 'access_level'] = 'Account Admin'
subset_df_system_access.tail()
```
